### Initialization of catalog

```python
%sql

CREATE CATALOG IF NOT EXISTS ecommerce
MANAGED LOCATION 'abfss://unity-catalog-data@amecommerceadls007.dfs.core.windows.net/ecommerce-catalog'
COMMENT 'E-COMMERCE PROJECT CATALOG CREATED IN EU-WEST, BACKED BY EXTERNAL LOCATION'
```

### Creating schemas for ecommerce catalog

```python
%sql

USE CATALOG ecommerce;

CREATE SCHEMA IF NOT EXISTS ecommerce.bronze;
CREATE SCHEMA IF NOT EXISTS ecommerce.silver;
CREATE SCHEMA IF NOT EXISTS ecommerce.gold;
```

```python
%sql
-- Create a managed volume in the raw schema for storing streaming checkpoints
CREATE VOLUME IF NOT EXISTS ecommerce.raw.checkpoints
COMMENT 'Checkpoint location for streaming jobs';
```
