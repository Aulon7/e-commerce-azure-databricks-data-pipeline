### Creating external volume for ADLS (`raw`)

```python
%sql
USE CATALOG ecommerce;

CREATE SCHEMA IF NOT EXISTS ecommerce.raw;

CREATE EXTERNAL VOLUME IF NOT EXISTS raw.raw_landing
  LOCATION 'abfss://ecommerce-raw-data@amecommerceadls007.dfs.core.windows.net/'
  COMMENT 'Landing zone for raw data'
```
